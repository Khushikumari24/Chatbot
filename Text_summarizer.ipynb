```python
from transformers import pipeline  #- Hugging Face provides ready-made NLP pipelines. ex- summerziation , translation
```

Load Summarization Model

```python
summarizer = pipeline(  #-Creates NLP workflow automatically.
    "summarization", #- tasktype
    model="facebook/bart-large-cnn"  #Pretrained transformer model from:
)
```

Input Large Text

```python
text = """
Artificial Intelligence is transforming the world.
AI is used in healthcare, education, robotics,
finance, and self-driving cars.
Machine learning and deep learning are subsets of AI.
Transformers have revolutionized NLP tasks.
"""
```

Generate Summary

```python
summary = summarizer(
    text,
    max_length=50,
    min_length=20,
    do_sample=False
)
print(summary[0]['summary_text'])
```
